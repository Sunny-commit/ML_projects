```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10 ,0.5)
# x=np.arange(intial point,final point ,no of steps to plot)
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('simple plot')
y=x**2
plt.plot(x,y,'r.')
```

```python
from pylab import *
plt.plot(x,sin(x),'k.')
plt.plot(x,cos(x),'b-')
```

```python
# object oriented apporach
fig=plt.figure()
```

```python
ax=fig.add_axes([0,0,1,1])
ax.set_title('sine wave')
ax.plot(x,y)
```

```python
ax.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,'ys-')
ax.plot(x,sin(x),'go--')
ax.set_xlabel('x-axis')
ax.set_ylabel('y_axis')
ax.set_title('Legend example')
ax.legend(labels=('tv','smartphone'),loc=('lower right'))
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.4,0.1,0.8,0.8])
ax2=fig.add_axes([0.55,0.55,0.3,0.3])
ax1.plot(x,y,'r.')
ax2.plot(x,sin(x),'g.')
ax2.set_title('sine wave')
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')
ax1.set_title('simple plot')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')
```

```python
fig,a=plt.subplots(2,2)
a[0][0].plot(x,y)
a[0][1].plot(x,sin(x))
a[1][0].plot(x,-sin(x))
a[1][1].plot(x,cos(x))
```

```python
# plt.subplot2grid(shape,location,rowspan,colspan)
a1=plt.subplot2grid((3,3),(0,0),colspan=2)
a2=plt.subplot2grid((3,3),(0,2),rowspan=33)
a3=plt.subplot2grid((3,3),(1,0),rowspan=2,colspan=2)
a1.plot(x,y)
a2.plot(x,sin(x))
a3.plot(x,cos(x))
```

Grid plots 

```python
fig,axes=plt.subplots(1,3,figsize=(12,4))
axes[0].plot(x,y)
axes[1].plot(x,sin(x))
axes[2].plot(x,cos(x))
axes[0].grid(True)
axes[1].grid(color='r',ls='-.',lw=0.25)
```

```python
# formatting axes
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
ax.set_xticklabels(['zero','two','four','six','eight','ten'])
```

```python
#Twin Axes
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
a2=ax.twinx()
a2.plot(x,y)
```

```python
# Bar plots
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=['c','c++','java','python','php']
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```
