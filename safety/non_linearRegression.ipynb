```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(-5.0, 5.0, 0.1)
print(x)
y=2*x+3
print(y)
```

```python
y_noise = 2 * np.random.normal(size=x.size)
print(y_noise)
```

# y=log(x)  

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = np.log(X)
plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```
