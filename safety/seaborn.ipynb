```python
import seaborn as sns
import matplotlib.pyplot as plt
c=sns.load_dataset("penguins") 
sns.barplot(x=c.island,y=c.bill_depth_mm,data=c,hue="sex",orient="v",color="b",palette="icefire",saturation=10,errcolor="y",errwidth=20,capsize=1,dodge=False,alpha=0.9)
plt.show()
```

```python
order_1=["Dream","Torgersen"]
sns.barplot(x=c.island,y=c.bill_depth_mm,data=c,order=order_1,hue_order=["female","male"])
plt.show()
```
