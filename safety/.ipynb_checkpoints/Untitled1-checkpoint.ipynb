```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(0,10,2)
y=[i**2 for i in x]
plt.plot(x,y)
print(x,y)
plt.title("simple line plot")
plt.xlabel("x-axis")
plt.ylabel("x squares")
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.plot([100,200,300,400])
plt.show()
```

```python
x=np.arange(1,10,2)
y=x**2
print(x,y)
plt.plot(x,y,marker="*",ms="12",mfc="r",mec="b")
plt.title("graph using marker,size,color")
plt.show()
```

```python
x=np.arange(0,10,2)
y=x**2
y1=x**3
print(x,y,y1)
plt.title("lines properties")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y1,marker="*",mfc="y")
plt.plot(x,y,y1,ls="dotted",lw=1,c="g")
plt.show()
```

```python
x=np.arange(1,10,2)
y=x*10
print(x,y)
f1={'family':'cambria','color':'red','size':14}
f2={'family':'cambria','color':'green','size':14}
plt.title("mutliple of x",fontdict=f1)
plt.xlabel("x-axis",fontdict=f2)
plt.ylabel("multiple of x",fontdict=f2)
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[20,40,50,60,78]
plt.plot(x,y)
# plt.grid(axis="x")
# plt.grid(axis="y")
plt.grid(axis="both",ls="dotted",color="r",lw="1")
plt.show()
```

```python
x=[1,2,3,5,6,]
y1=[50,60,40,20,67]
y2=[56,78,32,15,89]
plt.suptitle("multiple plots")
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("plot-1")
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("plot-2")
plt.subplot(2,2,3)
plt.plot(x,y1)
plt.title("plot-3")
plt.subplot(2,2,4)
plt.plot(x,y2)
plt.title("plot-4")
plt.show()
```

```python
x=np.arange(0,10,2)
y=x**2;
y1=x**3;
plt.plot(x,y)
plt.plot(x,y1)
plt.legend(["square","cubes"])
plt.plot(x,y,label="square")
plt.plot(x,y1,label="cubes")
plt.show()
```
