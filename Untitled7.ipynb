```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import math
```

```python
x=np.arange(1,10,0.25)
y=np.sin(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('simple plot')
plt.plot(x,y)
plt.show()
```

```python
# pylab module
from pylab import *
x=linspace(-3,3)
y=x**2
plot(x,y)
plt.show()
```

```python
plot(x,y,'r.')
plt.show()
```

```python
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x) 
plt.title('Sin angle')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plot(x,y)
plt.show()
```

```python
plot(x,sin(x))
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xlabel('angle')
ax.set_ylabel('sine')
ax.set_title('Sine wave')
ax.plot(x,y)
```

```python
y=[1,4,9,16,25,36,49,64]
x1=[1,16,30,42,55,68,77,88]
x2=[1,6,12,18,28,40,52,65]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
l1=ax.plot(x1,y,'ys-')
l2=ax.plot(x2,y,'go--')
ax.legend(labels=('tv','Smartestphone'),loc='lower right')
ax.set_title('Advertisement effect on sales')
ax.set_xlabel('median')
ax.set_ylabel('sales')
plt.show()
```
